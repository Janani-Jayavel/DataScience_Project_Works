```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder , MinMaxScaler, StandardScaler
```

```python
data['Item_Weight'].fillna(data['Item_Weight'].mean(),inplace = True)
```

```python
data['Outlet_Size'].fillna(data['Outlet_Size'].mode()[0],inplace = True)
```

```python
data['Item_Fat_Content'].replace({'Low Fat':0, 'low fat' : 0, 'LF':0 ,'Regular' :1, 'reg':1}, inplace= True)
```

```python
data['Outlet_Size'].replace({'Medium':1,'High':2,'Small':0},inplace = True)
```

```python
data['Outlet_Location_Type'].replace({'Tier 1':1,'Tier 2':2,'Tier 3':3},inplace = True)
```

```python
for i in data.select_dtypes(include = np.number).columns[:-1]:
    print(i)
    q1 = data[i].quantile(0.25)
    q3 = data[i].quantile(0.75)
    iqr = q3 - q1
    
    low = q1- 1.5*iqr
    hi = q3+ 1.5*iqr
    print("outlier =",data.loc[(data[i]<low) | (data[i]>hi),i])
```
